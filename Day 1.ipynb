```python
# Create simple DataFrame
Sample = [("iPhone", 999), ("Samsung", 799), ("MacBook", 1299)]
df = spark.createDataFrame(Sample, ["product", "price"])
df.show()

# Filter expensive products
df.filter(df.price > 1000).show()
```

```python
from pyspark.sql.functions import *

df2 = df.withColumn("category",when(df.price >= 1000, "Premium").otherwise("Standard"))

df2.show()
```
