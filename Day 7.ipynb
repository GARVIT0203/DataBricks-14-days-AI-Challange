```python

layer = dbutils.widgets.get("layer")

layer_paths = {
    "bronze": "workspace.bronze.events",
    "silver": "workspace.silver.events",
    "gold": "workspace.gold.events"
}

source_path = layer_paths[layer]
```

```python

def run_layer(layer):
    source_path = layer_paths[layer]

    if layer == "bronze":
        raw = spark.read.format("json").load(source_path)
        raw.write.mode("overwrite").saveAsTable(
            "workspace.bronze.events"
        )

    elif layer == "silver":
        df = spark.table("workspace.bronze.events")
        df_clean = df.dropDuplicates()
        df_clean.write.mode("overwrite").saveAsTable(
            "workspace.silver.events"
        )

    elif layer == "gold":
        df = spark.table("workspace.silver.events")
        agg = df.groupBy("event_type").count()
        agg.write.mode("overwrite").saveAsTable(
            "workspace.gold.events"
        )
```
