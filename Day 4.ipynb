```python
# Convert to Delta
events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv", header=True, inferSchema=True)
```

```python
%sql
SHOW VOLUMES IN workspace.ecommerce;

```

```python
events.write.format("delta").mode("overwrite").save(
    "/Volumes/workspace/ecommerce/ecommerce_data/events"
)
```

```python
# Create managed table
events.write.format("delta").saveAsTable("events_data")
```

```python

# SQL approach
spark.sql("""
    CREATE TABLE IF NOT EXISTS events_delta
    USING DELTA
    AS SELECT * FROM events_data
""")
```

```python
display(
    spark.sql("SELECT * FROM events_delta LIMIT 10")
)
```
